```python
!pip install yfinance

import yfinance as yf
import datetime
import requests
import pandas as pd
import numpy as np

start = datetime.datetime(2019, 3, 4)
end = datetime.datetime(2019, 12, 11)

# Fetch Daily data for MSCI World Index
msci_world = yf.download("^MXX", start=start, end=end, interval='1d')
msci_world.to_csv('MSCI World Index.csv')

# Fetch Daily data for S&P 500 Index
spx = yf.download("^GSPC", start=start, end=end, interval='1d')
spx.to_csv('S&P 500 Index.csv')

# Fetch Daily data for VIX Index
vix = yf.download("^VIX", start=start, end=end, interval='1d')
vix.to_csv('VIX Index.csv')

# Fetch Crude Oil WTI
wti = yf.download("CL=F", start=start, end=end, interval="1d")
wti.to_csv('Crude Oil.csv')

# Fetch EUR/USD
eur_usd = yf.download("EURUSD=X", start=start, end=end, interval="1d")
eur_usd.to_csv('EUR_USD.csv')

# Fetch CNY/USD
cny_usd = yf.download("CNY=X", start=start, end=end, interval="1d")
cny_usd.to_csv('CNY_USD.csv')

# Fetch JPY/USD
jpy_usd = yf.download("JPY=X", start=start, end=end, interval="1d")
jpy_usd.to_csv('JPY_USD.csv')

# Fetch Bitcoint data
btc_data = yf.download("BTC-USD", start=start, end=end, interval="1d")
btc_data.to_csv('Bitcoin.csv')

#Fetch Gold data
gold_data = yf.download('GC=F', start=start, end=end , interval="1d")
gold_data.to_csv('gold.csv')

##########################################
##########################################


#Fetch number of Transactions data
url = "https://api.blockchain.info/charts/n-transactions?timespan=41weeks&rollingAverage=1days&format=csv&start=2019-03-04"

response = requests.get(url)

num_trnc = pd.read_csv(url, header=0)

num_trnc = np.array(num_trnc)

num_trnc = pd.DataFrame({'Date': num_trnc[:,0], 'Value': num_trnc[:, 1]})

num_trnc['Date'] = pd.to_datetime(num_trnc['Date'])

num_trnc.to_csv('#Transaction.csv' , index= False)


#Fetch Mempool dataset
url2 = "https://api.blockchain.info/charts/mempool-size?timespan=41weeks&format=csv&start=2019-03-04"

mempool = pd.read_csv(url2 , header = 0)

mempool = pd.DataFrame(mempool)

mempool = np.array(mempool)

mempool = pd.DataFrame({'Date': mempool[:,0], 'Value': mempool[:, 1]})

mempool['Date'] = pd.to_datetime(mempool['Date'])

mempool = mempool.groupby(pd.Grouper(key='Date', freq='D')).mean()

mempool.to_csv('mempool.csv')
```
