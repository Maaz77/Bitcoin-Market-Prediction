```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('BTC-2019min.csv')

# Convert the 'Date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Filter the data between 4 March 2019 and 10 December 2019
start_date = pd.to_datetime('2019-03-04')
end_date = pd.to_datetime('2019-12-18')
filtered_data = df.loc[(df['date'] >= start_date) & (df['date'] <= end_date)]


filtered_data = filtered_data[['date','open','high','low','close']]

#inverse the rows,to have increasing date
filtered_data = filtered_data.iloc[::-1]

#drop the last row
filtered_data = filtered_data.drop(filtered_data.index[-1])


# Save the filtered data to a new CSV file
filtered_data.to_csv('bitcoin_minuntely.csv', index=False)

bitcoin = filtered_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

btc_data = bitcoin


btc_data = pd.DataFrame(btc_data)

# Create subplots with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(18, 10))

# Plot the Open price
axs[0, 0].plot(btc_data['open'])
axs[0, 0].set_title('Bitcoin Open Price')

# Plot the High price
axs[0, 1].plot(btc_data['high'])
axs[0, 1].set_title('Bitcoin High Price')

# Plot the Low price
axs[1, 0].plot(btc_data['low'])
axs[1, 0].set_title('Bitcoin Low Price')

# Plot the Close price
axs[1, 1].plot(btc_data['close'])
axs[1, 1].set_title('Bitcoin Close Price')

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plots
plt.show()
```

Here the returns median are calculated for the **classification** :

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

bitcoin = pd.read_csv('bitcoin_minuntely.csv')

bitcoin = bitcoin[['date','high']]

num_rows = bitcoin.shape[0]



m = [1, 5, 15, 60 , 10080]
returns_mean = []
allreturns = []

for horizon in m:

  returns = list()

  for i in range(int(num_rows * 5/9)):

    returns.append( (bitcoin['high'].iloc[ i + horizon ] / bitcoin['high'].iloc[i]) - 1 )
  
  allreturns.append(returns)

  #returns_mean.append(np.mean(returns))

np.shape(allreturns)
for i in range(len(allreturns)):
  print (np.mean(allreturns[i]))
```

```python
bitcoin_train = bitcoin[: int(num_rows * 5/9)]

bitcoint_val = bitcoin[int(num_rows * 5/9) : int (num_rows * 6/9)]

bitcoin_test = bitcoin[int (num_rows * 6/9) : ]


bitcoin_train.loc[: , '1m return'] = allreturns[0]
bitcoin_train.loc[: , '1m class'] = np.array(allreturns[0] < np.mean(allreturns[0])).astype(int)

bitcoin_train.loc[: , '5m return'] = allreturns[1]
bitcoin_train.loc[: , '5m class'] = np.array(allreturns[1] < np.mean(allreturns[1])).astype(int)

bitcoin_train.loc[: , '15m return'] = allreturns[2]
bitcoin_train.loc[: , '15m class'] = np.array(allreturns[2] < np.mean(allreturns[2])).astype(int)

bitcoin_train.loc[: , '60m return'] = allreturns[3]
bitcoin_train.loc[: , '60m class'] = np.array(allreturns[3] < np.mean(allreturns[3])).astype(int)

bitcoin_train.loc[: , '1week return'] = allreturns[4]
bitcoin_train.describe()

```

Following the feature values are standardized :

```python
for i in range (2 , len(bitcoin_train.columns) ,2):

  # Extract the 'return' columns from the dataset
  retrn_data = bitcoin_train.iloc[:, i].values.reshape(-1, 1)

  # Standardize the 'high' column using the StandardScaler from scikit-learn
  scaler = StandardScaler()
  retrn_data_standardized = scaler.fit_transform(retrn_data)

  # Replace the 'high' column in the dataset with the standardized values
  bitcoin_train[bitcoin_train.columns[i]] = retrn_data_standardized.flatten()


bitcoin_train.head()
```
